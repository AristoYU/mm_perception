```python
import os
import random

root_dir = '/projects/mm_perception'
os.chdir(root_dir)
```

```python
from mmengine.registry import DATASETS
from mmengine.config import Config

from mmdet.visualization import DetLocalVisualizer
```

```python
visualizer = DetLocalVisualizer()

config_file = 'configs/det2d/_base_/datasets/coco_detection.py'
cfg = Config.fromfile(config_file)
train_dataset = DATASETS.build(cfg.train_dataloader.dataset)
test_dataset = DATASETS.build(cfg.test_dataloader.dataset)
```

```python
train_idx = random.randint(0, len(train_dataset) - 1)

print(train_idx)
data = train_dataset[train_idx]
visualizer.add_datasample('train_img',
                          data['inputs'].permute(1, 2, 0).numpy()[:, :, ::-1],
                          data['data_samples'],
                          draw_pred=False)
visualizer.show()
```

```python
test_idx = random.randint(0, len(test_dataset) - 1)

print(test_idx)
data = test_dataset[test_idx]
visualizer.add_datasample('test_img',
                          data['inputs'].permute(1, 2, 0).numpy()[:, :, ::-1],
                          data['data_samples'],
                          draw_pred=False)
visualizer.show()
```
